```python
#import the required modules
import numpy as np
import pandas as pd
import seaborn as sn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from scipy import stats
```

```python
train_data = pd.read_csv("data.csv")
```

# Data Exploration

# Data processing

```python
# Encoding Categorical variables
train_data = pd.get_dummies(train_data, columns = ["diagnosis"], drop_first = True)
```

```python
train_data.head()
```

```python
correlation_matrix = train_data.corr()
correlation_matrix
```

```python
features = train_data.drop(columns = "diagnosis_M", axis = 1)
features = features.columns
```

# Data Visualization

```python
for feature in features:
    plt.hist(train_data[train_data["diagnosis_M"] == 1][feature], color = "green", label = "Malignant", alpha = 0.7, density = True)
    plt.hist(train_data[train_data["diagnosis_M"] == 0][feature], color = "red", label = "Benign", alpha = 0.7, density = True)
    plt.title(feature)
    plt.ylabel("Probability")
    plt.xlabel(feature)
    plt.legend()
    plt.show()
```

# Feature Engineering

```python

x_input = train_data[features]
y_output = train_data["diagnosis_M"]
```

```python
# Splitting the data into train and test(80% for train data and 20% for tes data)
x_train, x_test, y_train, y_test = train_test_split(x_input, y_output, test_size=0.2, random_state=42)
```

```python
# Scaling using Z-score normalization 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# Model selection and training (Random Forest Classifier)

```python
# Model Selection and Training
model = RandomForestClassifier(random_state=42)
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

# Hyperparameter Tuning using Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators" : [50, 100, 200],
    "max_depth" : [None, 10, 20, 30],
    "min_samples_split" : [2, 5, 10],
    "min_samples_leaf" : [1, 2, 4]
}

grid_search = GridSearchCV(estimator = model,
                          param_grid = param_grid,
                          cv = 5,
                          scoring = "accuracy",
                          n_jobs = 1
                          )
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```
